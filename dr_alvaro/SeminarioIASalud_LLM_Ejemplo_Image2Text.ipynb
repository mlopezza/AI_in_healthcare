```python
from transformers import pipeline

# Carga el pipeline: operación (image to text) y modelo (llava)
pipe = pipeline("image-text-to-text", model="llava-hf/llava-interleave-qwen-0.5b-hf") #bombshelll/ViT_BioMedBert_Captioning_ROCO") #


# Variables de entrada:
# -- Imagen
imagen = 'Tuberculosis-103.png'
# -- Prompt (pregunta)
pregunta = '¿Esta radiografía tiene Tuberculosis?'  #'¿De qué se trata la imagen?'
# -- Número de Tokens
Num_Tokens = '¿Esta radiografía tiene Tuberculosis?'  #'¿De qué se trata la imagen?'

# Define image and question
messages = [
    {
        "role": "user",
        "content": [
            {"type": "image", "image": imagen},
            {"type": "text", "text": pregunta},
        ],
    },
]
outputs = pipe(text=messages, max_new_tokens=Num_Tokens)
```

```python
print('Respuesta del Modelo: ', outputs[0]["generated_text"][1]['content'])
```
